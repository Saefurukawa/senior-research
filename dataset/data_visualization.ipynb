```python
#data visualization
import json
import matplotlib.pyplot as plt

# Load the dataset from a JSON file
with open("easy_train_data.json", "r") as f:
    easy_train_data = json.load(f)

# Load the dataset from a JSON file
with open("easy_eval_data.json", "r") as f:
    easy_eval_data = json.load(f)
```

```python
#Label distribution  -> 0: hate speech 1: normal 2:offensive
# Define label mapping
label_mapping = {0: "Hate Speech", 1: "Normal", 2: "Offensive"}


label_train_counts ={}
for data in easy_train_data:
    label = data["label"]
    if label in label_train_counts:
        label_train_counts[label] +=1
    else:
        label_train_counts[label] =1
print(label_train_counts)

label_eval_counts ={}
for data in easy_eval_data:
    label = data["label"]
    if label in label_eval_counts:
        label_eval_counts[label] +=1
    else:
        label_eval_counts[label] =1
print(label_eval_counts)


# Convert numeric labels to text labels using the mapping
label_train_text = {label_mapping[key]: value for key, value in label_train_counts.items()}
label_eval_text = {label_mapping[key]: value for key, value in label_eval_counts.items()}

# Plotting the label distribution for training and evaluation data
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot for training data
ax1.bar(label_train_text.keys(), label_train_counts.values(), color='blue')
ax1.set_title("Training Data Label Distribution")
ax1.set_xlabel("Labels")
ax1.set_ylabel("Counts")

# Plot for evaluation data
ax2.bar(label_eval_text.keys(), label_eval_counts.values(), color='green')
ax2.set_title("Evaluation Data Label Distribution")
ax2.set_xlabel("Labels")
ax2.set_ylabel("Counts")

plt.tight_layout()
plt.show()
```

```python
target_train_counts ={}
for data in easy_train_data:
    targets = data['target']
    # print(targets)
    for target in targets:
        target = target.lower()
        if target not in target_train_counts:
            target_train_counts[target] = 0
        target_train_counts[target] +=1
print(target_train_counts)

target_eval_counts ={}
for data in easy_eval_data:
    targets = data['target']
    # print(targets)
    for target in targets:
        target = target.lower()
        if target not in target_eval_counts:
            target_eval_counts[target] =0
        target_eval_counts[target] +=1
print(target_eval_counts)

# Colorblind-friendly colors
colors = ['#D55E00', '#0072B2']

# Plotting the label distribution for training and evaluation data
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot for training data
ax1.bar(target_train_counts.keys(), target_train_counts.values(), color=colors[0])
ax1.set_title("Training Data Label Distribution", fontsize=14)
ax1.set_xlabel("Target Groups", fontsize=14)
ax1.set_ylabel("Counts", fontsize=14)
ax1.set_xticks(range(len(target_train_counts)))
ax1.set_xticklabels(target_train_counts.keys(), rotation=90, fontsize=10)

# Plot for evaluation data
ax2.bar(target_eval_counts.keys(), target_eval_counts.values(), color=colors[1])
ax2.set_title("Evaluation Data Label Distribution", fontsize=14)
ax2.set_xlabel("Target Groups", fontsize=14)
ax2.set_ylabel("Counts", fontsize=14)
ax2.set_xticks(range(len(target_eval_counts)))
ax2.set_xticklabels(target_eval_counts.keys(), rotation=90, fontsize=10)

plt.tight_layout()
plt.savefig("label_distribution.png", dpi=300, bbox_inches='tight')
plt.show()
```
